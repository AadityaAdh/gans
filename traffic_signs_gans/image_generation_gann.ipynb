# Image generation using Previously Trained GAN model

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from PIL import Image
```

```python
keep_right_model=load_model('models/keep_right_model_generator.h5')
no_entry_model=load_model('models/no_entry_model_generator.h5')
no_stop_model=load_model('models/no_stop_model_generator.h5')
```

```python
def rescale_images(images):
    return (images + 1) / 2
```

```python
while(True):
    user_prompt=input("enter what kind of image you want to generate \n")
    if "generate"  in user_prompt and "keep"  in user_prompt and "right"  in user_prompt:
        imgs = keep_right_model.predict(tf.random.normal((16, 100, 1)))

        img_rescaled = rescale_images(imgs)

        fig, ax = plt.subplots(ncols=4, nrows=4, figsize=(10,10))
        for r in range(4):
            for c in range(4):
                ax[r][c].imshow(img_rescaled[(r+1)*(c+1)-1])
        plt.savefig('tempimg.png')
        plt.close(fig)
        img = Image.open('tempimg.png')
        plt.imshow(img)
        plt.axis('off')  # Hide axes if not needed
        plt.show()





    elif "generate"  in user_prompt and "no"  in user_prompt and "entry"  in user_prompt:
        imgs = no_entry_model.predict(tf.random.normal((16, 100, 1)))

        img_rescaled = rescale_images(imgs)

        fig, ax = plt.subplots(ncols=4, nrows=4, figsize=(10,10))
        for r in range(4):
            for c in range(4):
                ax[r][c].imshow(img_rescaled[(r+1)*(c+1)-1])
        plt.savefig('tempimg.png')
        plt.close(fig)
        img = Image.open('tempimg.png')
        plt.imshow(img)
        plt.axis('off')  # Hide axes if not needed
        plt.show()





    elif "generate"  in user_prompt and "no"  in user_prompt and "stop"  in user_prompt:
        imgs = no_stop_model.predict(tf.random.normal((16, 100, 1)))

        img_rescaled = rescale_images(imgs)

        fig, ax = plt.subplots(ncols=4, nrows=4, figsize=(10,10))
        for r in range(4):
            for c in range(4):
                ax[r][c].imshow(img_rescaled[(r+1)*(c+1)-1])
        plt.savefig('tempimg.png')
        plt.close(fig)
        img = Image.open('tempimg.png')
        plt.imshow(img)
        plt.axis('off')  # Hide axes if not needed
        plt.show()


    elif "exit" in user_prompt:
        break



    else:
        print("Didn't Understand waht  you were saying can you try that again?")
    
    
```
